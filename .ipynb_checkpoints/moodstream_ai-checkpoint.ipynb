# Data Pre-Processing

```python
def preprocess_pixels(pixel_data):
    pixels = np.array(pixel_data.split(), dtype='float32').reshape(48, 48)
    pixels = np.stack([pixels] * 3, axis=-1)  
    pixels /= 255.0  # Normalize pixel values to [0, 1]
    return pixels
```

```python
label_to_int = {v: k for k, v in label_emotions.items()}

def data_generator(df, batch_size):
    while True:
        for start in range(0, len(df), batch_size):
            end = min(start + batch_size, len(df))
            batch_df = df[start:end]
            images = np.array([preprocess_pixels(pixels) for pixels in batch_df['pixels']])
            labels = batch_df['emotion'].map(label_to_int).values
            labels = to_categorical(labels, num_classes=7)
            yield images, labels
```

```python
train_df, valid_df = train_test_split(df, test_size=0.1, stratify=df['emotion'], random_state=42)
batch_size = 32
train_generator = data_generator(train_df, batch_size)
valid_generator = data_generator(valid_df, batch_size)
```

```python
from tensorflow.keras import Input

model = Sequential([
    Input(shape=(48, 48, 3)),  # Correctly specify the input shape as (48, 48, 3) for RGB images
    Conv2D(32, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(128, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(256, (3, 3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(pool_size=(2, 2)),

    GlobalAveragePooling2D(),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(7, activation='softmax')  
])

model.compile(optimizer=optimizers.AdamW(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])

# Early stopping and learning rate reduction
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6)

# Training model using the custom generators
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_df) // batch_size,
    validation_data=valid_generator,
    validation_steps=len(valid_df) // batch_size,
    epochs=50,
    callbacks=[early_stopping, reduce_lr]
)
```

```python
loss, accuracy = model.evaluate(valid_generator, steps=len(valid_df) // batch_size)
print(f'Validation Accuracy: {accuracy}')
```

```python
def build_model(bottom_model, classes):
    model = bottom_model.layers[-2].output
    model = GlobalAveragePooling2D()(model)
    model = Dense(classes, activation = 'softmax', name = 'out_layer')(model)
    
    return model
```
